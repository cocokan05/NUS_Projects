```python
class convers_pca():
    def __init__(self, no_of_components):
        self.no_of_components = no_of_components
        self.eigen_values = None
        self.eigen_vectors = None
        
    def transform(self, x):
        return np.dot(x - self.mean, self.projection_matrix.T)
    
    def inverse_transform(self, x):
        return np.dot(x, self.projection_matrix) + self.mean
    
    def fit(self, x):
        self.no_of_components = x.shape[1] if self.no_of_components is None else self.no_of_components
        self.mean = np.mean(x, axis=0)
        
        cov_matrix = np.cov(x - self.mean, rowvar=False)
        
        self.eigen_values, self.eigen_vectors = np.linalg.eig(cov_matrix)
        self.eigen_vectors = self.eigen_vectors.T
        
        self.sorted_components = np.argsort(self.eigen_values)[::-1]
           
```

```python
pca = convers_pca(no_of_components=10)
pca.fit(data)
```

```python
pca.eigen_vectors
```

```python
kmeans6 = KMeans(6)
kmeans6.fit(data_p_t)
```

```python
identified_clusters6 = kmeans6.fit_predict(data_p_t)
identified_clusters6
```

```python
data_with_clusters = data_p_t.copy()
data_with_clusters['Clusters'] = identified_clusters6
data_with_clusters['r']=r
data_with_clusters['v']=v
data_with_clusters.tail()
```

```python
plt.scatter(data_with_clusters['r'],data_with_clusters['v'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.title("K Means (k=6)")
plt.xlabel('annual return')
plt.ylabel('volatility')
plt.show()
```
